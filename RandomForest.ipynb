Use famous iris flower dataset from sklearn.datasets to predict flower species using random forest classifier.

Measure prediction score using default n_estimators (10)
Now fine tune your model by changing number of trees in your classifer and tell me what best score you can get using how many trees

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. Load iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# 2. Split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Default Random Forest with 10 trees
model_default = RandomForestClassifier(n_estimators=10, random_state=42)
model_default.fit(X_train, y_train)
y_pred_default = model_default.predict(X_test)
default_accuracy = accuracy_score(y_test, y_pred_default)

print(f"Default (n_estimators=10) Accuracy: {default_accuracy:.4f}")

# 4. Fine-tuning: try different numbers of trees
best_score = 0
best_trees = 0

for trees in range(5, 201, 5):  # try trees from 5 to 200
    model = RandomForestClassifier(n_estimators=trees, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = accuracy_score(y_test, y_pred)

    if score > best_score:
        best_score = score
        best_trees = trees

print(f"Best Accuracy: {best_score:.4f} with n_estimators={best_trees}")
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Load breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target   # 0 = malignant (cancer), 1 = benign (no cancer)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Accuracy
print("âœ… Accuracy:", accuracy_score(y_test, y_pred))

# Classification report (precision, recall, f1-score)
print("\nðŸ“Š Classification Report:\n", classification_report(y_test, y_pred, target_names=data.target_names))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=data.target_names,
            yticklabels=data.target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Breast Cancer Detection")
plt.show()
```
