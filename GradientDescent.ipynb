```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 5, 6])   # somewhat linear but with noise

# Parameters
m = 0   # slope
b = 0   # intercept
L = 0.01   # learning rate
epochs = 1000   # iterations
n = len(x)

# Store history for visualization
m_history = []
b_history = []
loss_history = []

# Gradient Descent
for i in range(epochs):
    y_pred = m * x + b
    error = y - y_pred

    # Compute gradients
    dm = -(2/n) * sum(x * error)
    db = -(2/n) * sum(error)

    # Update parameters
    m = m - L * dm
    b = b - L * db

    # Compute mean squared error
    loss = np.mean(error**2)

    # Save for later
    m_history.append(m)
    b_history.append(b)
    loss_history.append(loss)

# Print final values
print(f"Final slope (m): {m:.4f}")
print(f"Final intercept (b): {b:.4f}")
print(f"Final loss: {loss:.4f}")

# Plot data and best-fit line
plt.scatter(x, y, color="red", label="Data points")
plt.plot(x, m*x+b, color="green", label="Best-fit line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression using Gradient Descent")
plt.show()

# Plot loss curve
plt.plot(loss_history)
plt.xlabel("Iteration")
plt.ylabel("Loss (MSE)")
plt.title("Error Reduction Over Time")
plt.show()
```

 Example 1: Fit a Line (Linear Regression)

We already saw this, but lets keep a cleaner version:

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 5, 6])

# Initialize parameters
m, b = 0, 0
alpha = 0.01   # learning rate
epochs = 1000
n = len(x)

# Gradient Descent
for _ in range(epochs):
    y_pred = m * x + b
    error = y - y_pred
    m -= alpha * (-(2/n) * sum(x * error))
    b -= alpha * (-(2/n) * sum(error))

print(f"Line: y = {m:.2f}x + {b:.2f}")

plt.scatter(x, y, color="red")
plt.plot(x, m*x+b, color="blue")
plt.show()
```

 Example 2: Find Square Root by Minimization

We can use gradient descent to approximate a by minimizing

()=(2)2


```python
a = 25   # we want sqrt(25) = 5
x = 1.0  # initial guess
alpha = 0.01

for i in range(1000):
    grad = 4*x*(x**2 - a)   # derivative of (x^2 - a)^2
    x -= alpha * grad

print(f"Approximated sqrt({a}) = {x:.4f}")
```

 Example 3: Minimize a Quadratic Function

We minimize:

f(w)=w
2
+2w+1

which has a known minimum at w = -1.

```python
w = 10.0    # start far away
alpha = 0.1

for i in range(50):
    grad = 2*w + 2   # derivative of w^2 + 2w + 1
    w -= alpha * grad
    if i % 5 == 0:
        print(f"Iteration {i}: w = {w:.4f}")

print(f"Minimum found at w = {w:.4f}")
```

 Example 4: Logistic Regression (Classification)

Using gradient descent to separate 2D points:

```python
import numpy as np
import matplotlib.pyplot as plt

# Simple dataset (AND logic)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

# Initialize weights
w = np.zeros(X.shape[1])
b = 0
alpha = 0.1
epochs = 1000

# Sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Gradient Descent
for _ in range(epochs):
    z = np.dot(X, w) + b
    y_pred = sigmoid(z)
    error = y_pred - y
    w -= alpha * np.dot(X.T, error) / len(y)
    b -= alpha * np.sum(error) / len(y)

print("Weights:", w, "Bias:", b)

# Predictions
print("Predictions:", (sigmoid(np.dot(X, w) + b) > 0.5).astype(int))
```
